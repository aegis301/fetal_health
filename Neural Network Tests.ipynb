# Neural Network

This notebook is used to do some simple tests on neural networks for this data set.

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
```

```python
data = pd.read_csv('data/fetal_health.csv')
data.head()
```

```python
data.info()
```

```python
features = data.drop(['fetal_health'], axis=1)
labels = data['fetal_health']

features = np.array(features)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=808)
```

```python
normalize = layers.Normalization()
normalize.adapt(X_train)


easy_model = tf.keras.Sequential([
    normalize,
    layers.Dense(265, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(3)
])

easy_model.compile(loss = tf.keras.losses.MeanAbsoluteError() ,optimizer=tf.optimizers.Adam(), metrics=["accuracy"])
easy_model.fit(X_train, y_train, epochs=50)
easy_model.evaluate(X_test, y_test)
```
